```python
import pandas as pd
import numpy as np
import os
from sqlalchemy import create_engine
import sys
os.getcwd()
sys.path.append("C:/SQL_PW/")
import config 

path = "//Resource//"
file = os.path.join("Resource", "Food_Access_Research_Atlas.csv")
FoodAccess_df=pd.read_csv(file)
FoodAccess_df.head()
```

```python
# Get seperate Urban and Rural data where there is no Food Quarter in the region
#While individuals living in these areas may appear to be far from a supermarket or grocery store, 
#they may not truly experience difficulty accessing healthy and affordable food.
data_urban=FoodAccess_df[(FoodAccess_df["Urban"]==1) & (FoodAccess_df["NUMGQTRS"]==0)]
data_rural=FoodAccess_df[(FoodAccess_df["Rural"]==1) & (FoodAccess_df["NUMGQTRS"]==0)]
```

```python
#Get tracts with Low income and low access measured at 1/2 and 10 miles in urban area
data_urban_1And10=data_urban[(data_urban["LILATracts_1And10"]==1) & 
                             (data_urban["LILATracts_halfAnd10"]==0) & 
                             (data_urban["LILATracts_1And20"]==0)]

#Get tract with Low income and low access measured at 1 and 10 miles in urban area
data_urban_1And10=data_urban[(data_urban["LILATracts_halfAnd10"]==1) & 
                             (data_urban["LILATracts_1And20"]==0)]


```

```python
#Get tract with Low income and low access measured at 1 and 10 miles in Rural area
data_rural_1And10=data_rural[(data_rural["LILATracts_halfAnd10"]==1) & 
                             (data_rural["LILATracts_1And20"]==0)]

#Get tract with Low income and low access measured at 1 and 20 miles in Rural area
data_rural_1And10=data_rural[(data_rural["LILATracts_1And20"]==1)]
```
